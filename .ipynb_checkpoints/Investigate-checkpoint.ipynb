```python
# plot correlation heatmap
plt.rcParams['figure.figsize'] = [20, 10]
Data_frame_correlation = data.corr()
sn.heatmap(Data_frame_correlation, annot=True);
```

```python
Data_frame_correlation
```

```python
Matrix_corr=[]
Matrix_corr_features=[]
Rows_names_Corr = Data_frame_correlation.index

for i in range (0,len(Data_frame_correlation.columns)):
    
    for j in range (0,len(Data_frame_correlation.columns)):
        
        if Data_frame_correlation[Data_frame_correlation.columns[i]][j]>0 and Data_frame_correlation[Data_frame_correlation.columns[i]][j]!=1 :
            
            print('correlation beteween',Data_frame_correlation.columns[i],
                  
                  '|and|',
                  
                  Rows_names_Corr[j],
                  
                  '=',
                  
                  Data_frame_correlation[Data_frame_correlation.columns[i]][j])
            Matrix_corr.append([Data_frame_correlation.columns[i],Rows_names_Corr[j],Data_frame_correlation[Data_frame_correlation.columns[i]][j]])
            Matrix_corr_features.append([Data_frame_correlation.columns[i],Rows_names_Corr[j]])
```

```python
Matrix_corr_features
```

```python
unique_features_corr=np.unique(Matrix_corr_features)
unique_features_corr
print('Length of unique_features_correlation',len(unique_features_corr))
print(unique_features_corr)
```

```python
indices = np.where(unique_features_corr=='id')
unique_features_corr = np.delete(unique_features_corr, indices)
unique_features_corr
```

```python
data.columns
```

```python
index=0

while index < len(unique_features_corr) :
    for i in range ( index, len(unique_features_corr)):
        if i == index:
            pass
        else:

            data.plot(x=unique_features_corr[i] ,y=unique_features_corr[index],
                    
                    kind='scatter',
                    fontsize=15,
                    subplots=True,
                    sharex=True
                    );
            plt.title(unique_features_corr[index] +' VS ' + unique_features_corr[i]  ,fontsize=20)
            plt.xlabel(unique_features_corr[i], fontsize=20)
            plt.ylabel(unique_features_corr[index] , fontsize=20)
    index+=1
        
```

```python
years=data['release_year'].unique()
years
```

```python
years=sorted(years,reverse=True)
```

```python
len(years)
```

```python
profit_per_year=[]
No_of_movies_per_year=[]

for i in range ( 0 , len(years)):
    prof=data[data['release_year']==years[i]]
    total=prof['Profit'].sum()
    count=prof['release_year'].count()
    profit_per_year.append(total)
    No_of_movies_per_year.append(count)
```

```python
len(profit_per_year)
```

```python
labels= years
number_of_samples=profit_per_year
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
ax.bar(labels,number_of_samples)
plt.gca().invert_xaxis()
plt.title('Total profit earned per year in the movie industry from 1961 to 2015'  ,fontsize=30)
plt.xlabel('Release Year', fontsize=30)
plt.ylabel('Total profit per Year' , fontsize=30)
plt.grid()
plt.show()
```

```python
labels= years
number_of_samples=No_of_movies_per_year
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
ax.bar(labels,number_of_samples)
plt.gca().invert_xaxis()
plt.title('Total number of movies produced in the industry from 1961 to 2015'  ,fontsize=30)
plt.xlabel('Release Year', fontsize=30)
plt.ylabel('number_of_movies_per_year' , fontsize=30)
plt.grid()
plt.show()
```

```python
caegories=data['Categories']
caegories=np.array(caegories)
```

```python
final_categories=[]

for i in range (0,len(caegories)):
    
    x=caegories[i].split("|")
    
    final_categories.append(x)

final_categories
```

```python
data['Categories_splitred']=final_categories
```

```python
data.head()
```

```python
final_categories[0]
```

```python
unique_categories=[]
for i in range ( 0,len(final_categories)):
    for category in final_categories[i]:
        if category not in unique_categories:
            unique_categories.append(category)
unique_categories            
```

```python
len(unique_categories)
```

```python
length=[]
for i in range (0,len(unique_categories)):
    plot_category=data[data['Categories'].str.contains(unique_categories[i])]
    fig=plt.figure()
    length.append(len(plot_category))
    plt.title(unique_categories[i]+' Movies Production from 1961 to 2015',fontsize=20)
    plot_category['release_year'].value_counts().plot(kind='bar')
```
