# Dataset Description

In this section we will provide a very detailed description for each Feature provided in the dataset

**TMDb movie data description**: 

> 1- TMDb is a Movies dataset consists of **10866 Rows** (Records) and **19 columns** (Features)

> 2- TMDb dataset collected from the Movie Database which includes Millions of movies, TV shows and people to discover

> 3- This dataset is cleaned from the original dataset published on [Kaggle](https://www.kaggle.com/tmdb/tmdb-movie-metadata)

> 4- TMDB dataset describes different important aspects and points about each movie in order to help the stake holder and the analytic's team to explore and extract important informations and patterns that helps in different fields such as recommnedation systems and more.

> 5- TMDB dataset contains **19 columns** Features summrized as follows :- 
    >> **ID** :- The Entry (Movie) information Number each Entry (Row) takes unique number in the cleaned dataset from **Kaggle** . 
        >>> Datatype:- integer
    
    >> **imdb_id** :- The Entry (Movie) information Number each Entry (Row) takes unique number in the original TMDB **database**. 
        >>> Datatype:- integer
       
    >> **popularity** :- Its a number indicates how much the Movie is popular, a **higher** value means a **more** popular, a **lower** value indicates a **Less** popular.
        >>> Datatype:- Float
        
    >> **budget** :- **approximate** number or expenses describes the total amount of money spent on the Movie ( from the begining of the production until the first release and show in the market.
            >>> Datatype:- Integer
 
    >> **Revenue** :- **approximate** number or income describes the total amount of money taken from the movie selling from different places such as ( cinema shows, post office and more)
            >>> Datatype:- Integer
            
    >> **original_title** :- its a setence or even a sigle word descibes the title of the movie
            >>> Datatype:- String
            
     >> **cast** :- is the group of actors names who make up the Movie or stage play, each record have more than one name separated with "|"  
            >>> Datatype:- String

     >> **homepage** :- displays the full website link for each movie 
            >>> Datatype:- String
            
     >> **director** :- is the group of poeple names who make up the movie production motage and other staff to make out the movie with its final version, each record have more than one name separated with "|"
            >>> Datatype:- String
            
     >> **tagline** :- is a short text which serves to clarify an idea for as well as to give a little bit description about the movie title, some times is a motivated sentence to give impression of suspense about the movie. 
            >>> Datatype:- String
            
     >> **keywords** :- A keyword is a words to describe any notable object, concept, style or action that takes place during a title. each keyword is separated with "|"
            >>> Datatype:- String
            
     >> **overview** :- A one or more sentence describe and represent a breif about the movie story and how thes story starts.
            >>> Datatype:- String
            
     >> **runtime** :- a number represents the overall duration time of the movie this number in **minutes**
            >>> Datatype:- Integer
        
     >> **genres** :- represents the category of the movie if its action or romantic movie or maybe with multiple categories. each category separated by "|"
            >>> Datatype:- String
            
     >> **production_companies** :- The companies names that responsable on the production process as well as the expenses. maybe will be one or more than one company. Each company separated by "|"
            >>> Datatype:- String 
            
     >> **release_date** :- The date that the movies released in. the format is "DD/MM/YYY"
            >>> Datatype:- String
            
     >> **vote_count** :- how many ppl votes and rates the movie
            >>> Datatype:- Integer   
            
    >> **budget_adj** :- **Extact** number or expenses describes the total amount of money spent on the Movie ( from the begining of the production until the first release and show in the market.
            >>> Datatype:- Float
 
    >> **revenue_adj** :- **Extact** number or income describes the total amount of money taken from the movie selling from different places such as ( cinema shows, post office and more)
            >>> Datatype:- Float

#### Research Question 1 :- is the Movie industry worth it to invest in it ?

```python
years=df['release_year'].unique()
years=sorted(years,reverse=True)
years
```

```python
profit_2009=df[df['release_year'] == 2015]
profit_2009['release_year'].count()
```

```python
profit_per_year=[]
number_of_movies_per_year=[]
for i in range ( 0 , len (years)):
    prof=df[df['release_year'] == years[i]]
    total=prof['Profit'].sum()
    count=prof['release_year'].count()
    profit_per_year.append(total)
    number_of_movies_per_year.append(count)
    total=0
    count=0
```

```python
profit_per_year
```

```python


labels = years
number_of_samples = profit_per_year

fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax.bar(labels, number_of_samples)
plt.gca().invert_xaxis()
plt.title('Total profit earned per year in the movie industry from 1961 to 2015'  ,fontsize=30)
plt.xlabel('Release Year', fontsize=30)
plt.ylabel('Total profit per Year' , fontsize=30)
plt.grid()
plt.show()
```

```python

labels = years
number_of_samples = number_of_movies_per_year

fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax.bar(labels, number_of_samples)
plt.gca().invert_xaxis()
plt.title('Total number of movies produced in the industry from 1961 to 2015'  ,fontsize=30)
plt.xlabel('Release Year', fontsize=30)
plt.ylabel('number_of_movies_per_year' , fontsize=30)
plt.grid()
plt.show()
```

```python
years=df['release_year'].unique()
years
```

```python
categories=df['categories']
categories=np.array(categories)
len(categories)
```

#### Finding the Unqiue Categories of the movies in order to plot each category if exits in category columns Vs Year

```python
final_category=[]
for i in range (0,len(categories)):
        x = categories[i].split("|")
        final_category.append(x)

final_category
# print(len(final_category))
```

```python
print(len(df))
```

```python
df['Categories_splitted']=final_category
df
```

```python
unique_categories=[]
for i in range ( 0, len(final_category)):
    
    for category in final_category[i]:
        if category not in unique_categories:
            unique_categories.append(category)
unique_categories
```

```python
lengths=[]
for i in range ( 0, len (unique_categories )):
    plot_category= df[df['categories'].str.contains(unique_categories[i])]
    fig = plt.figure()
    lengths.append(len(plot_category))
    plt.title(unique_categories[i]+' Movies Production from 1961 to 2011',fontsize=20)
    plot_category['release_year'].value_counts().plot(kind='bar')
```

## Observation 

We can observe from the 12 figures that all the movies for any category produced **exponentaly increased** from 1961 to 2015.


```python
# Training Set
import matplotlib
import matplotlib.pyplot as plt
import numpy as np


labels = unique_categories
number_of_samples = lengths

x = np.arange(len(labels))  # the label locations
width = 0.35  # the width of the bars

fig, ax = plt.subplots()
rects1 = ax.bar(labels, number_of_samples, width)

ax.set_xlabel('Categorty',fontsize=20)
ax.set_ylabel('Number of Samples',fontsize=20)
ax.set_title('Total No of Movies produced for each category from 1961 to 2015',fontsize=20)
ax.set_xticks(x)
ax.set_xticklabels(labels)
ax.legend()
def autolabel(rects):
    for rect in rects:
        height = rect.get_height()
        ax.annotate('{}'.format(height),xy=(rect.get_x() + rect.get_width() / 2, height),xytext=(0, 1),  # 3 points vertical offset
                    textcoords="offset points",
                    ha='center', va='bottom')
autolabel(rects1)
fig.tight_layout()
plt.grid()
plt.show()
```
