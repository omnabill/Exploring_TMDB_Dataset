### Q1 : Which is the most popular actor/actress in the cinema history ?

### to answer this question we must start splitting cast as cast of each film is stored at 'cast' field but delimeted by '|' 
  >> **first we create a dictionary of each actor and cout of his apperance
  
  >> **second we placed this dictionary at a dataframe to start visualize it data
  
  >> **lastly we will visualize it's data using bar plot

```python
#df['cast'][42]
```

```python
my_dict={}
for i in range(len(df)):
    for l in df['cast'][i].split('|'):
            if l not in my_dict:
                        my_dict[l]=1
            else:
                        my_dict[l] = my_dict.get(l, 0) + 1


```

```python
my_dict
```

```python
Actors = pd.DataFrame.from_dict(my_dict, orient='index', columns=['Apperance_count'])
```

```python
Actors.head()
```

```python
Actors.sort_values(by='Apperance_count', ascending=False, inplace=True)
```

```python
Actors.head(10)
```

```python
plt.figure(figsize=(14,7))

# Add title
plt.title("Top 10 Most Apperance Actors",fontsize=14)
plt.ylabel('Count the apperance',fontsize=14)
plt.xticks(rotation = 70,fontsize=10,fontstyle='italic')
sns.barplot(x=Actors.index[0:20], y=Actors['Apperance_count'][:20])
```

## So Let's Answer our third question 

### Q3 : Which companies are the Top profited companies over the years ?


### we will fill a dictionary of companies mapped with their profits as their some movies which may be 2 companies or more cooperated at their production, so i divided the profits of such films over the number of the companies, compained this dictionary to a dataframe then plotting the top most 10 profitted companies as pie chart plot .

```python
comp_dict={}
for i in range(len(df)):
    for l in df['production_companies'][i].split('|'):
            if l not in comp_dict:
                        comp_dict[l]=(df['profit'][i]/len(df['production_companies'][i].split('|')))
            else:
                        comp_dict[l]+= (df['profit'][i]/len(df['production_companies'][i].split('|')))


```

```python
comp_dict
```

```python
profit = pd.DataFrame.from_dict(comp_dict, orient='index', columns=['profit'])
```

```python
profit.head()
```

```python
profit.sort_values(by='profit', ascending=False, inplace=True)
```

```python
profit.head(10)
```

```python
profit = profit.head(10)
```

```python
maxi = 0 
for l in profit['profit']:
    if l > maxi:
        maxi= l
```

```python
labels = profit.index
profits = profit['profit']
explode = [0.4 if i == maxi else 0 for i in profit['profit']]  # explode highest company
fig1, ax1 = plt.subplots()
ax1.pie(profits, explode=explode, labels=labels, autopct='%1.1f%%',
        shadow=True, startangle=10,radius=5)
ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

plt.show()

```
